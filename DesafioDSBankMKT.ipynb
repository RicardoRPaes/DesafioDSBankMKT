```python
#%% 1 - Qual profissão tem mais tendência a fazer um empréstimo? De qual tipo?

# dataframe contagem dos 'job' e classe de 'housing'
prep1 = df0.groupby(['job','housing']).size().reset_index().rename(columns={0:'cont_total'})

# dataframes de contagem housing == 'yes' e housing == 'no'
prep1_yes = prep1[(prep1.housing == 'yes')].reset_index(drop=True).drop(['housing'], axis=1).rename(columns={'cont_total': 'cont_housing_yes'})
prep1_no = prep1[(prep1.housing == 'no')].reset_index(drop=True).drop(['housing'], axis=1).rename(columns={'cont_total': 'cont_housing_no'})

# dataframe para plot
df1 = prep1.groupby(['job'], as_index=False).sum()
df1 = df1.sort_values(['cont_total'], ascending=[False]).reset_index(drop=True)
df1 = pd.merge(df1, prep1_yes, how='outer', on='job')
df1 = pd.merge(df1, prep1_no, how='outer', on='job')

# plot gráfico 1
data = [go.Bar(x=df1.job,
            y=df1.cont_housing_yes,name='para habitação'),
        go.Bar(x=df1.job,
            y=df1.cont_housing_no,name='outros')
       ]
layout = go.Layout(
    yaxis=dict(title='número de empréstimos'),
    title='Tipos de Empréstimos por Profissão', barmode='stack'
)
fig = go.Figure(data=data, layout=layout)

iplot(fig,filename='gráfico 1')
```

```python
#%% 2 - Fazendo uma relação entre número de contatos e sucesso da campanha quais são os pontos relevantes a serem observados?

# número de adesões antes da campanha
numAdesAntCamp = ((df0['y'] == 'yes') & (df0['campaign'] == 0)).sum()

# número de adesões
numAdes = (df0['y'] == 'yes').sum()

# número total de contatos realizados antes e durante a campanha
numContatos = (df0['campaign'].sum() + df0['previous'].sum())

# taxa de conversão da campanha
txConv = round((100 * numAdes / numContatos), 2)

# número mínimo de contatos realizados durante a campanha
numMinCont = min(df0['campaign'])

print('número de adesões antes da campanha = ' + str(numAdesAntCamp))
print('\nnúmero de adesões = ' + str(numAdes))
print('\nnúmero de contatos realizados = ' + str(numContatos))
print('\ntaxa de conversão = ' + str(txConv) + '%')
print('\nnúmero mínimo de contatos realizados durante a campanha = ' + str(numMinCont))
```

# 2 - Fazendo uma relação entre número de contatos e sucesso da campanha quais são os pontos relevantes a serem observados?

Antes da campanha ser realizada, o número de clientes que assinam o depósito a prazo (y = no) é zero. A adesão dos clientes começa a ocorrer apenas durante a campanha, visto que é necessário ao menos 1 contato para que o y do cliente seja equivalente a ‘yes’. Durante as campanhas, o número total de adesões (y = yes) é de 5.289, somando-se todos os contatos realizados (antes e durante a campanha) foi obtido um total de 151.193 ligações, ou seja, uma taxa de conversão de apenas 3.5% (5.289 / 151.193).

```python
#%% 3 - Baseando-se nos resultados de adesão desta campanha qual o número médio e o máximo de ligações que você indica 
#   para otimizar a adesão?

# dataframe 3 para plot
df3 = df0[df0['y'] == 'yes'].groupby(['campaign', 'y']).size().reset_index().rename(columns={0:'cont_y_yes'})
df3 = df3.sort_values(['cont_y_yes'], ascending=[False]).reset_index(drop=True)

df3['share_contatos_Suces'] = round(100 * (df3['cont_y_yes'] / numAdes), 2).cumsum()

# plot gráfico 3
trace1 = go.Bar(
    x=df3.campaign,
    y=df3.cont_y_yes,
    name='número de adesões por contato'
)
trace2 = go.Scatter(
    x=df3.campaign,
    y=df3.share_contatos_Suces,
    name='share acumulado de adesões',
    yaxis='y2'
)

data = [trace1, trace2]

layout = go.Layout(
    xaxis=dict(title='número de ligações realizadas'),
    yaxis=dict(title='número adesões'),
    yaxis2=dict(title='share (%)', overlaying='y', side='right'),
    title='Otimização de Adesão',
    legend=dict(x=-.1, y=1.2)
)
fig = go.Figure(data=data, layout=layout)

iplot(fig,filename='gráfico 3')
```

# 3 - Baseando-se nos resultados de adesão desta campanha qual o número médio e o máximo de ligações que você indica para otimizar a adesão?

Na figura acima, as colunas azuis representam o número de adesões obtidas cada quantidade de ligações observadas na amostra. Já a linha laranja, representa o share do número de adesões obtidos por cada quantidade de ligações com relação ao total de adesões obtidas em toda amostra (100 x nº de adesões de cada quantidade de ligação / nº total de adesões da amostra). 
Sob o ponto de vista do princípio de Pareto, onde aproximadamente 20% dos esforços deve representar aproximadamente 80% dos resultados, afirmar que o número otimizado médio de ligações deve 3 (86,59% na linha laranja) é uma aproximação bastante razoável.
Analisando ainda mais o gráfico acima, pode-se notar que o número máximo para otimização de adesões deve ser 11 (99,15% na linha laranja). Acima de 11 ligações, as adesões passam a ser insignificantes.

```python
#%% 4 -O resultado da campanha anterior tem relevância na campanha atual?

# dataframe contagem dos 'poutcome' e classe de 'y'
prep4 = df0.groupby(['poutcome','y']).size().reset_index().rename(columns={0:'cont_total'})

# dataframes de contagem y == 'yes' e y == 'no'
prep4_yes = prep4[(prep4.y == 'yes')].reset_index(drop=True).drop(['y'], axis=1).rename(columns={'cont_total': 'cont_y_yes'})
prep4_no = prep4[(prep4.y == 'no')].reset_index(drop=True).drop(['y'], axis=1).rename(columns={'cont_total': 'cont_y_no'})

# dataframe 4 para plot
df4 = prep4.groupby(['poutcome'], as_index=False).sum()
df4 = df4.sort_values(['cont_total'], ascending=[False]).reset_index(drop=True)
df4 = pd.merge(df4, prep4_yes, how='outer', on='poutcome')
df4 = pd.merge(df4, prep4_no, how='outer', on='poutcome')

# taxa de conversão com relação às campanhas anteriores
df4['tx_conv_camp_ant'] = round((100 * df4['cont_y_yes'] / df4['cont_total']), 2)

# plot gráfico 4

trace1 = go.Bar(
    x=df4.poutcome,
    y=df4.cont_y_yes,
    name='nº de sucessos (campanha atual)',
)

trace2 = go.Bar(
    x=df4.poutcome,
    y=df4.cont_y_no,
    name='nº de fracassos (campanha atual)',
)

trace3 = go.Scatter(
    x=df4.poutcome,
    y=df4.tx_conv_camp_ant,
    name='share de sucesso da campanha atual',
    yaxis='y2'
)

data = [trace1, trace2, trace3]

layout = go.Layout(
    xaxis=dict(title='status da campanha anterior'),
    yaxis=dict(title='contagem'),
    yaxis2=dict(title='share (%)', overlaying='y', side='right'),
    title='Comparativo: Campanha Atual x Campanha Anterior', barmode='stack',
    legend=dict(x=-.1, y=1.0, orientation='h')
)
fig = go.Figure(data=data, layout=layout)

iplot(fig,filename='gráfico 4')
```

# 4 - O resultado da campanha anterior tem relevância na campanha atual?

Para cada status da campanha anterior, no gráfico acima, as colunas laranjas representam laranjas o número de fracassos ocorridos na campanha atual (y = no), as colunas azuis referem-se ao número de sucessos obtidos na campanha atual e a linha verde representa o share de sucessos da campanha atual (nº de sucessos / (nº de sucessos + nº de fracassos)). Como pode ser observado, o share de sucesso no status success (64,73%) tende a ser preponderantemente maior que os demais status (unknown, failure, other), os quais não superam a marca de 17%. Este é um indício de que a adesão tende a ser mais efetiva com clientes que já aderiram às campanhas passadas.

```python
#%% 5 - Qual o fator determinante para que o banco exija um seguro de crédito?

# dataframe para ser encodado: trocar valor str por int
prep5 = df0.copy()

toEncoder5 = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']

from sklearn.preprocessing import LabelEncoder

for i in toEncoder5:
    prep5[i] = LabelEncoder().fit_transform(prep5[i])

# plot gráfico 5a
df5a = prep5.corr(method='spearman')

fig, ax = plt.subplots()
sb.heatmap(df5a, vmin=-1, vmax=1.0, cmap='RdBu_r', robust=False, annot=True)
#plt.show()
ax.set_title('Correlação Entre as Variáveis (Coeficiente de Spearman)')
fig.set_size_inches(17.55, 12.405)

# plot gráfico 5b
df5b = prep5[['default','balance']].copy()

trace1 = go.Box(y=df5b[df5b.default==1].balance,
    name='balance para default = yes',
    boxmean=True,
    boxpoints=False,
    yaxis='y'
)
trace2=go.Box(y=df5b[df5b.default==0].balance,             
    name='balance para default = no',
    boxmean=True,
    boxpoints = False,
    yaxis='y'
)

data = [trace1, trace2]

layout = go.Layout(
    title='Boxplot: Balance x Default',
    yaxis = dict(
        anchor='y',
        title='balance',
        range=[-500,2000]
    )
)
    

fig = go.Figure(data=data, layout=layout)

iplot(fig,filename='gráfico 5b')
```

# 5 - Qual o fator determinante para que o banco exija um seguro de crédito?

O mapa de calor (gráfico acima) é a representação da correlação entre cada uma das variáveis contidas no dataset fornecido. Para esta análise, foi utilizado o coeficiente de correlação de Spearman. Ao observar a coluna da variável default, nota-se que, com exceção da própria variável default, a variável que possui maior correlação com default é a balance (-0.17, uma correlação negativa). 
Levando em conta que balance refere-se à média do saldo anual do cliente, consequentemente, eventuais saldos negativos devem impactar bastante em decisões de análise de crédito. Esta hipótese pode ser comprovada através do boxplots gerados com base nos dados de balance e default dos clientes do banco. Para default = yes, 75% dos clientes possuem balance que variam entre -340 e 15 Euros. Já para default = no, 75% encontra-se com balance variando entre 84 e 1462,5 Euros.

```python
#%% 6 - Quais são as características mais proeminentes de um cliente que possua empréstimo imobiliário?

# considerando apenas a variáveis que não possuem ligação com campanhas
prep6 = df0[['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan']]

toEncoder6 = ['job', 'marital', 'education', 'default', 'housing', 'loan']

for i in toEncoder6:
    prep6[i] = LabelEncoder().fit_transform(prep6[i])
    
# plot gráfico 6a
df6a = prep6.corr(method='spearman')

fig, ax = plt.subplots()
sb.heatmap(df6a, vmin=-1, vmax=1.0, cmap='RdBu_r', robust=False, annot=True)
#plt.show()
ax.set_title('Correlação Entre as Variáveis Sem Ligação com Campanha (Coeficiente de Spearman)')
fig.set_size_inches(17.55, 12.405)

# plot gráfico 6b
df6b = prep6[['age','housing']].copy()

trace1 = go.Box(y=df6b[df6b.housing==1].age,
    name='com housing loan',
    boxmean=True,
    boxpoints=False,
    yaxis='y'
)
trace2=go.Box(y=df6b[df6b.housing==0].age,             
    name='sem housing loan',
    boxmean=True,
    boxpoints=False,
    yaxis='y'
)

data = [trace1, trace2]

layout = go.Layout(
    title='Boxplot: Age x Housing Loan',
    yaxis=dict(
        anchor='y',
        title='age',
        range=[10,100]
    )
)

fig = go.Figure(data=data, layout=layout)

iplot(fig,filename='gráfico 6b')
```
